### Univariate analysis on numerical variables

```python
plt.figure(figsize = (20,10))
ax1 = sns.boxplot(x = "age", data = df)
plt.grid(True)
plt.xlabel("Age", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.show()

plt.figure(figsize = (20,10))
plt.grid(True)
ax1 = sns.boxplot(x = "bmi", data = df)
plt.xlabel("BMI", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.show()

plt.figure(figsize = (20,10))
plt.grid(True)
ax1 = sns.boxplot(x = "children", data = df)
plt.xlabel("Children", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.show()

plt.figure(figsize = (20,10))
plt.grid(True)
ax1 = sns.boxplot(x = "charges", data = df)
plt.xlabel("Charges", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.show()
```

The above output gives the information about the numerical variables in the data set graphically. We have chosen the boxplot to represent the numerical data as the boxplot gives the minimum , 1st quartile(25%) , median(50%) , 3rd quartile(75%) and maximum values of that column or variable

Age : As we can see 50% of the insured are in the age group of 38 to 40 where as the minimum and maximum age as per the data are 18 and 62-62 respectively. As a whole 25% to 75% insured are between the age group of 27 to 52 .

BMI : The ideal BMI value lies between 18.5 and 24.9 . According to the the data set 50% of the insured lie around the value of 31 to 32 which is not ideal . 25% to 75% of the insured lie around the values 26 to 34 . There are a handful of insured whose BMI is above the value 46 . 

Children : 25% of the insured do not have a child where 50% of them have one child and 75% of them have 2 children . Rest of the insured have 3 , 4 , or 5 children.

Charges : 25% of the charges lie around 5000 dollars where as 50% and 75% of them lie around 10000 and 15000 dollars respectively . The rest either lie around 1000 to 5000 dollars and 15000 to 35000 dollars . There are very less number of insured whose charges lie above 35000 dollars which might be due to their known medical conditions .

The wider the width of the violin the higher frequency of the corresponding value and the narrower the width lower the frequency of the corresponding value .

The number of male insured is more compred to that of female in the region of high charges as more men travel around for work when compared to women and hence they can be the ones prone to more diseases.

Smoker v/s Charges : The charges start from 10000 dollars for that of smoker and the charges lie within 10000 dollars for most non smokers which implies smoking is an important feature for an insured to decide on the charges by the company. As we can see the width of the violin varies as the charges increases which can be due to the age of the insured and the number of family members or children . Most of the number of charges of that of the smokers lie between 20000 to 40000 dollars where as for non smokers it lies between 5000 to 12000 dollars. The width the of the violin of the smokers narrows from around 60000  and ends at around 72000 dollars, as for that of non smokers the violin narrows from 30000 and ends at around 40000 dollars. 

Region v/s Charges : 50% of the insured living in different regions of the US correspond to the charges around 10000 dollars. The width of the violin start getting narrow and then again widens around 30000 to 50000 for southeast, southwest and northeast regions mostly . The charges are higher for southeast region when compared to the other regions, this might be due to higher cost of living in that region .

## Linear Regression Model

### Using Standardization

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X = df2[["age", "sex", "bmi", "children", "smoker", "region"]]
y = df2["charges"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =27)
```

```python
X_train_stand = X_train.copy()
X_test_stand = X_test.copy()

num_cols = ["age", "bmi", "children"]

for i in num_cols:
    scale = StandardScaler().fit(X_train_stand[[i]])

    X_train_stand[i] = scale.transform(X_train_stand[[i]])

    X_test_stand[i] = scale.transform(X_test_stand[[i]])

print(X_train_stand[0:5])
print(X_test_stand[0:5])

```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train_stand , y_train)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test_stand)

# mean_squared_error
mqe = np.sqrt(mean_squared_error(y_test , y_pred))

print(mqe)
```

```python
df3 = pd.DataFrame({"Actual" : y_test, "Predicted" : y_pred})
print(df3.head())
```
