## Linear Regression Model

```python
from scipy import stats
import statsmodels

data = pd.DataFrame(df2)
```

```python
X= data[["sex", "children","bmi", "smoker"]]
X["sex"] = np.log(X["sex"])
X["bmi"] = np.log(X["bmi"])

y = data["charges"]
y = np.log(y)
```

```python
print(X.head())
print(y.head())
```

```python
import statsmodels.api as smf

reg = smf.OLS(y, X).fit()


reg.summary()
```

#### The presence of correlation in error terms reduces models accuracy.If the error terms are correlated, the estimated standard errors tend to underestimate the true standard error.

#### Also, lower standard errors would cause the associated p-values to be lower than actual. This will make us incorrectly conclude a parameter to be statistically significant.

#### Look for Durbin  Watson (DW) statistic in the above output. It must lie between 0 and 4. If DW = 2, implies no autocorrelation, 0 < DW < 2 implies positive autocorrelation while 2 < DW < 4 indicates negative autocorrelation.

#### As we can see the DW statistic is aproximately 2 which means there is no presence of autocorrelation.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 


vif_data = pd.DataFrame() 
vif_data["feature"] = X.columns 
#### There are two major things we can notice:
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                          for i in range(len(X.columns))] 

print(vif_data) 
```

#### Multicollinearity phenomenon exists when the independent variables are found to be moderately or highly correlated. In a model with correlated variables, it becomes a tough task to find out which variable is actually contributing to predict the response variable.

#### With presence of correlated predictors, the standard errors tend to increase. Also, when predictors are correlated, the estimated regression coefficient of a correlated variable depends on which other predictors are available in the model. If this happens, we will end up with an incorrect conclusion that a variable strongly / weakly affects target variable. Even if you drop one correlated variable from the model, its estimated regression coefficients would change.

#### We can use VIF(Variance inflation factor) factor. VIF value <= 4 suggests no multicollinearity whereas a value of >= 10 implies serious multicollinearity. 

#### As we can see in the above output none of the variables are correlated and hence multicollinearity phenomenon doe not exits in model built on the dataset.

```python
pred_val = reg.fittedvalues.copy()
true_val = y.copy()
residual = true_val - pred_val

print(residual.head())

print(residual.tail())
```

```python
#### There are two major things we can notice:
plot_lm_1 = plt.figure(figsize = (20,10))
plot_lm_1.axes[0] = sns.residplot(data.columns[-1], residual, data=data,
                          lowess=True,
                          scatter_kws={'alpha': 0.5},
                          line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})

plot_lm_1.axes[0].set_title('Residuals vs Fitted')
plot_lm_1.axes[0].set_xlabel('Fitted values')
plot_lm_1.axes[0].set_ylabel('Residuals')
```

#### As we can see the above the plot, we can consider it as the sign of constant variance i.e. homoskedasticity. Also,we can use Breusch-Pagan / Cook  Weisberg test or White general test to detect this phenomenon.

#### To overcome heteroskedasticity, a possible way is to transform the response variable such as log(Y) or Y. Also, we can use weighted least square method to tackle heteroskedasticity.

```python
from statsmodels.graphics.gofplots import qqplot
import matplotlib as mpl

with mpl.rc_context():
    mpl.rc("figure", figsize=(20,10))
    qqplot(residual, line = 'r', alpha = 0.5)
```

#### As we can see in the above qq(quantile-quantile) plot there is normality in the distribution met and hence the model validates the normality assumption.

#### If the errors are not normally distributed, non  linear transformation of the variables (response or predictors) can bring improvement in the model.
