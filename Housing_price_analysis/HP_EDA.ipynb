```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv("train.csv")
df.head()
df.info()
params = {'axes.titlesize':'30',
          'xtick.labelsize':'20',
          'ytick.labelsize':'20'}
plt.rcParams.update(params)
df.hist(figsize = (50,50), bins = 100);
plt.savefig("f5.png")
plt.show()
```

```python
df2 = df[[column for column in df if df[column].count() / len(df) > 0.3]]
del(df2["Id"])
print("List of dropped columns ")
for column in df:
    if column not in df2:
        print(column + " ")
print("\n")
```

```python
df2.info()
params = {'axes.titlesize':'30',
          'xtick.labelsize':'20',
          'ytick.labelsize':'20'}
plt.rcParams.update(params)
df2.hist(figsize = (50,50), bins = 100);
plt.savefig("f4.png")
plt.show()
```

```python
df.head()
```

```python
params = {'axes.titlesize':'25',
          'xtick.labelsize':'20',
          'ytick.labelsize':'20'}
plt.rcParams.update(params)
plt.figure(figsize = (10,10))
sns.distplot(df2["SalePrice"], bins = 100, kde = True, hist_kws = { 'color' : 'r','alpha' : 0.4, 'histtype' : 'step', 'fill' : True}, kde_kws = {'color' : 'b', 'alpha' : 0.5});
plt.xlabel("Sale Price", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.savefig("f1.png")
plt.show()
```

```python
df_num = df2.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

```python
params = {'axes.titlesize':'30',
          'xtick.labelsize':'20',
          'ytick.labelsize':'20'}
plt.rcParams.update(params)
df_num.hist(figsize = (50,50), bins = 100);
plt.savefig("f2.png")
plt.show()


```

```python
print(df_num.describe())
params = {'axes.titlesize':'15',
          'xtick.labelsize':'10',
          'ytick.labelsize':'10'}
plt.rcParams.update(params)
plt.figure(figsize = (50,50))
for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num,
                x_vars=df_num.columns[i:i+5],
                y_vars=['SalePrice'])
```

```python
corr = df2.corr()
params = {'axes.titlesize':'25',
          'xtick.labelsize':'30',
          'ytick.labelsize':'30'}
plt.rcParams.update(params)
plt.figure(figsize = (50, 50))
sns.heatmap(corr [(corr >= 0.5) | (corr <= -0.4)], 
           xticklabels=corr.columns.values, 
           yticklabels=corr.columns.values, 
           cmap="viridis",
          annot=True, annot_kws = {'fontsize' : 18})
plt.savefig("f6.png")

```

```python
cor = df2.corr()
cor_sp = cor["SalePrice"]
print(cor_sp)
final_cor = pd.DataFrame(cor_sp[cor_sp > 0.5].dropna())
print(final_cor)
quant_list = [column for column in df_num if column in final_cor.index]
fig, ax = plt.subplots(round(len(quant_list) / 3), 3 , figsize = (25,20))
params = {'axes.titlesize':'15',
          'xtick.labelsize':'8',
          'ytick.labelsize':'8'}
plt.rcParams.update(params)
for i, ax in enumerate(fig.axes):
    if i < len(quant_list) :
        sns.regplot(x = quant_list[i],y = 'SalePrice', data = df2[quant_list], ax=ax)
```

```python
cat_list = [column for column in quant_list[:-1] + df2.columns.tolist() if (column not in quant_list) or (column not in df2.columns.tolist())]
df_cat_list = df2[cat_list]
print(df_cat_list.head())
df_not_num = df_cat_list.select_dtypes(include = 'O')
print(df_not_num.columns)
```

```python
plt.figure(figsize = (10,10))
params = {'axes.titlesize':'15',
          'xtick.labelsize':'8',
          'ytick.labelsize':'8'}
plt.rcParams.update(params)
ax = sns.boxplot(x = "PavedDrive", y = "SalePrice", data = df2)
plt.setp(ax.artists, alpha = 0.5, linewidth = 2, edgecolor = "k")
plt.xticks(rotation = 45)
plt.ylabel("Sale Price")
plt.xlabel("Paved Drive")
plt.savefig("f01.png")
plt.show()
```

```python
plt.figure(figsize = (10,10))
ax = sns.boxplot(x = "SaleCondition", y = "SalePrice", data = df2)
plt.setp(ax.artists, alpha = 0.5, linewidth = 2, edgecolor = "k")
plt.xticks(rotation = 45)
plt.ylabel("Sale Pice")
plt.xlabel("Sale Condition")
plt.savefig("f02.png")
plt.show()
```
